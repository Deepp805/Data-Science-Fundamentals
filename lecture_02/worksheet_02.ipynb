### Effective Programming

a) What is a drawback of the top down approach?

One drawback is that we will focus on the smaller individual parts and forget to consider the whole problem we are supposed to be focused on. Another is that it can be harder to make sure what we have done is correct until we actually finish our implementation and test on real inputs.

b) What is a drawback of the bottom up approach?

We spend a lot of the initial time focusing on the approach rather than implementation. It may also result in more code being written overall, some of which may be unused in the final result

c) What are 3 things you can do to have a better debugging experience?

1. Read the error message. Look at what the error message is saying and where the issue occurred in the code
2. Read the code carefully. Walk through the code in your head or on paper to see where the error happens and why
3. Seek external help through online sources

```python
class Board:
    def __init__(self):
        self.board = [["_" for _ in range(8)] for _ in range(8)]
        self.solutions = int()
    
    def __repr__(self):
        res = ""
        for row in self.board:
            for col in row:
                res += col
                res += " "
            res += "\n"
        return res

    def is_legal_row(self, row):
        for c in range(len(self.board[0])):
            if self.board[row][c] == "Q":
                return False
        return True
    
    def is_legal_col(self, col):
        for r in range(len(self.board)):
            if self.board[r][col] == "Q":
                return False
        return True
    
    def is_on_board(self, row, col):
        return row >= 0 and row < 8 and col >= 0 and col < 8

    def is_legal_diag(self, row, col):
        for i in range(len(self.board)):
            if self.is_on_board(row -i, col-i) and self.board[row-i][col-i] == "Q":
                return False
            if self.is_on_board(row +i, col+i) and self.board[row+i][col+i] == "Q":
                return False
            if self.is_on_board(row -i, col+i) and self.board[row-i][col+i] == "Q":
                return False
            if self.is_on_board(row +i, col-i) and self.board[row+i][col-i] == "Q":
                return False

        return True

    def is_legal(self, row, col):
        if not self.is_legal_row(row):
            return False
        
        if not self.is_legal_col(col):
            return False
        
        if not self.is_legal_diag(row, col):
            return False
        
        return True

    def set_queen_at(self, row, col):
        self.board[row][col] = "Q"

    def get_queen_on(self, row):
        for c in range(len(self.board[0])):
            if self.board[row][c] == "Q":
                return c
        raise ValueError("Programmer Error for finding a queen")

    def unset_queen_on(self, row):
        self.board[row] = ["_" for _ in range(8)]
    
    def search(self):
        row = 0
        col = 0
        self.solutions = 0

        while row >= 0:
            if row < 8:
                if self.is_legal(row, col):
                    self.set_queen_at(row, col)
                    row += 1
                    col = 0
                else:
                    col += 1
                    while col >= 8:
                        if row == 0:
                            return                        
                        col = self.get_queen_on(row-1)
                        self.unset_queen_on(row-1)
                        col += 1
                        row -= 1
            else:
                print("found solution")
                print(self)
                self.solutions += 1
                col = self.get_queen_on(row-1)
                self.unset_queen_on(row-1)
                col += 1
                row -= 1
                if row == 0:
                    return
                elif col >= 8:
                    col = self.get_queen_on(row-1)
                    self.unset_queen_on(row-1)
                    col += 1
                    row -= 1
                    if row == 0:
                        return

my_board = Board()
my_board.search()
print(my_board.solutions)
```
